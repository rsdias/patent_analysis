```python
#November 09, 2021
#Assignee e' um dado categorico, para uma regressao nao importa quem e'
#Localizacao tb

#September 30, 2021
#Estes sao os arquivos relevantes para assignee e location
#data/location.tsv.zip contem dados da localidade (city, country devem ser os mais relevantes)
#data/assignee.tsv.zip relaciona inventor (quando e' individuo) ou assignee
#data/patent_assignee.tsv.zip faz o de-para entre patentes, locations e assignees

# podemos trabalhar com id ou com os dados em si
# vale criar uma dummy para diferenciar assignee de inventor solo
# trabalhar com id diminui esforco computacional

# por ora, trabalhar somente com data/patent_assignee.tsv.zip
# introduzir as outras tabelas a posteriori

# para efeito do modelo estatistico, o dado de patent_assignee e' suficiente
# porem agregar o local permite fazer calculos de distancia
import pandas as pd
```

```python
location='data/location.tsv.zip'
df=pd.read_csv(location, compression='zip', sep='\t')
```

```python
df.head()
```

```python
location='data/patent_assignee.tsv.zip'
df2=pd.read_csv(location, compression='zip', sep='\t', dtype=object)
```

```python
df2.head()
```

```python
df=df2.merge(df, left_on='location_id', right_on='id', how='inner')
```

```python
df=df[['patent_id','assignee_id', 'location_id', 'city', 'state', 'country']].set_index('patent_id')
dst='data/pat_assignee.parquet.gz'
df = df.rename(columns={'country': 'country_asg'})
#df.to_parquet(dst, compression='gzip')
df.memory_usage()
```

```python
top_50=df.groupby('assignee_id').size().sort_values(ascending=False).head(50).to_frame().rename(columns={0:'number_patents'})
```

```python
asg='data/assignee.tsv.zip'
df_asg=pd.read_csv(asg, compression='zip', sep='\t')
df_asg=df_asg.set_index('id')
```

```python
df_asg
```

```python
top_50=top_50.join(df_asg, how='inner')
```

```python
from IPython.display import display, HTML

top_50=top_50.reset_index()[['organization', 'number_patents']]
#display(HTML(top_50.to_html()))
```

```python
#top 10 per decade
# distribution of foreign citations per year
patent='data/cleanpatent.parquet.gz'
df_pat = pd.read_parquet(patent)

df_pat=df_pat.merge(df2, right_on='patent_id', left_index=True, how='inner')
```

```python
#number of patents per year by assignee
top_10=df_pat.groupby(['date','assignee_id']).size()
```

```python
#top 10 by year
top_10=top_10.reset_index().sort_values(0, ascending=False).groupby('date').head(10).sort_values('date')
```

```python
top_10=top_10.merge(df_asg, left_on='assignee_id', right_index=True, how='inner')
```

```python
top_10=top_10.sort_values(['date', 0], ascending=False)[['date','organization',0 ]]
```

```python
#cria um ranking baseado no numero de patentes
#a manha aqui e' selecionar 'first' no metodo rank, evitando ranks repetidos em caso de empate
# e permitindo o pivot table
top_10['rank']=top_10.groupby("date")[0].rank("first", ascending=False)
```

```python
top_10=top_10.pivot(index='date', columns="rank", values='organization')
display(HTML(top_10.to_html()))
```

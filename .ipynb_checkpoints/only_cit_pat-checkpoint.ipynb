```python
dst='data/only_cit_pat.csv'

fp = open(dst, 'w')#abre o arquivo dst com permissão para escrita
writer = csv.writer(fp, delimiter=',')
headers=['patent_id', 'citation_id']
writer.writerow(headers)#a primeira linha recebe o nome da variavel
for chunk in df.get_chunk():
    for rows in chunk:
        row=[]
        for x in rows:
            if x:
                row.append(x.encode('utf-8'))
            else:
                row.append("")
#         print row
        writer.writerow(row)
fp.close()
```
